```python
# 簡單 Python 計算機

# 取得使用者輸入
num1 = float(input("請輸入第一個數字: "))
num2 = float(input("請輸入第二個數字: "))

# 選擇運算方式
print("選擇運算方式：")
print("1. 加法 (+)")
print("2. 減法 (-)")
print("3. 乘法 (*)")
print("4. 除法 (/)")
print("5. 整數除法 (//)")
print("6. 取餘數 (%)")
print("7. 次方 (**)")
choice = input("請輸入數字 1-7: ")

# 計算並顯示結果
if choice == "1":
    print("結果:", num1 + num2)
elif choice == "2":
    print("結果:", num1 - num2)
elif choice == "3":
    print("結果:", num1 * num2)
elif choice == "4":
    if num2 != 0:
        print("結果:", num1 / num2)
    else:
        print("錯誤：除數不能為 0")
elif choice == "5":
    if num2 != 0:
        print("結果:", num1 // num2)
    else:
        print("錯誤：除數不能為 0")
elif choice == "6":
    if num2 != 0:
        print("結果:", num1 % num2)
    else:
        print("錯誤：除數不能為 0")
elif choice == "7":
    print("結果:", num1 ** num2)
else:
    print("錯誤：無效選擇")
```
