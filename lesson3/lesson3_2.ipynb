```python
# BMI 身體質量指數計算程式
# BMI = 體重(kg) / 身高(m)²

print("=" * 30)
print("BMI 計算器")
print("=" * 30)

# 輸入身高（公分）
height_cm = float(input("請輸入身高（公分）："))

# 輸入體重（公斤）
weight_kg = float(input("請輸入體重（公斤）："))

# 將公分轉換為公尺
height_m = height_cm / 100

# 計算 BMI
bmi = weight_kg / (height_m ** 2)

# 顯示計算結果
print("=" * 30)
print(f"身高：{height_cm} 公分")
print(f"體重：{weight_kg} 公斤")
print(f"BMI：{bmi:.2f}")

# 判斷健康狀態
if bmi < 18.5:
    status = "體重過輕"
elif bmi < 24:
    status = "正常範圍"
elif bmi < 27:
    status = "過重"
else:
    status = "肥胖"

print(f"健康狀態：{status}")
print("=" * 30)
```
