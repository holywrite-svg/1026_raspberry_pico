```python
import paho.mqtt.client as mqtt
import time
import json
from datetime import datetime
```

```python
# MQTT Broker è¨­å®šï¼ˆèˆ‡ lesson6_1.ipynb ç›¸åŒï¼‰
MQTT_BROKER = "localhost"  # æˆ–ä½¿ç”¨ "192.168.0.252" å¦‚æœæ˜¯é ç«¯
MQTT_PORT = 1883
MQTT_USERNAME = "pi"
MQTT_PASSWORD = "raspberry"
MQTT_CLIENT_ID = "test_subscriber"

print(f"MQTT Broker: {MQTT_BROKER}:{MQTT_PORT}")
print(f"Username: {MQTT_USERNAME}")
```

```python
# å»ºç«‹ MQTT å®¢æˆ¶ç«¯ï¼ˆä½¿ç”¨ VERSION2 APIï¼‰
try:
    client = mqtt.Client(mqtt.CallbackAPIVersion.VERSION2, client_id=MQTT_CLIENT_ID)
    use_v2 = True
except AttributeError:
    # å¦‚æœç‰ˆæœ¬è¼ƒèˆŠï¼Œä½¿ç”¨èˆŠçš„æ–¹å¼
    client = mqtt.Client(client_id=MQTT_CLIENT_ID)
    use_v2 = False

# è¨­å®šä½¿ç”¨è€…åç¨±å’Œå¯†ç¢¼
if MQTT_USERNAME and MQTT_PASSWORD:
    client.username_pw_set(MQTT_USERNAME, MQTT_PASSWORD)

# é€£ç·šå›èª¿å‡½æ•¸ï¼ˆæ ¹æ“š API ç‰ˆæœ¬ä½¿ç”¨ä¸åŒçš„ç°½åï¼‰
if use_v2:
    def on_connect(client, userdata, flags, reason_code, properties):
        if reason_code == 0:
            print("âœ… æˆåŠŸé€£æ¥åˆ° MQTT Broker!")
        else:
            print(f"âŒ é€£ç·šå¤±æ•—ï¼ŒéŒ¯èª¤ä»£ç¢¼: {reason_code}")
else:
    def on_connect(client, userdata, flags, rc):
        if rc == 0:
            print("âœ… æˆåŠŸé€£æ¥åˆ° MQTT Broker!")
        else:
            print(f"âŒ é€£ç·šå¤±æ•—ï¼ŒéŒ¯èª¤ä»£ç¢¼: {rc}")

# è¨‚é–±å›èª¿å‡½æ•¸
def on_subscribe(client, userdata, mid, granted_qos, properties=None):
    print(f"âœ… æˆåŠŸè¨‚é–±ä¸»é¡Œ (Message ID: {mid}, QoS: {granted_qos})")

# æ¥æ”¶è¨Šæ¯å›èª¿å‡½æ•¸
def on_message(client, userdata, msg):
    timestamp = datetime.now().strftime("%H:%M:%S")
    print(f"\nğŸ“¨ [{timestamp}] æ”¶åˆ°è¨Šæ¯:")
    print(f"   ä¸»é¡Œ: {msg.topic}")
    print(f"   QoS: {msg.qos}")
    try:
        # å˜—è©¦è§£æ JSON
        data = json.loads(msg.payload.decode())
        print(f"   å…§å®¹ (JSON): {json.dumps(data, ensure_ascii=False, indent=2)}")
    except:
        # å¦‚æœä¸æ˜¯ JSONï¼Œç›´æ¥é¡¯ç¤ºæ–‡å­—
        print(f"   å…§å®¹: {msg.payload.decode()}")

# è¨­å®šå›èª¿å‡½æ•¸
client.on_connect = on_connect
client.on_subscribe = on_subscribe
client.on_message = on_message

# å¦‚æœå·²ç¶“é€£ç·šï¼Œå…ˆæ–·é–‹
try:
    client.loop_stop()
    client.disconnect()
except:
    pass

# é€£æ¥åˆ° Broker
try:
    client.connect(MQTT_BROKER, MQTT_PORT, 60)
    client.loop_start()  # åœ¨èƒŒæ™¯åŸ·è¡Œç¶²è·¯è¿´åœˆ
    time.sleep(1)  # ç­‰å¾…é€£ç·šå»ºç«‹
except Exception as e:
    print(f"âŒ é€£ç·šéŒ¯èª¤: {e}")
```

```python
# è¨‚é–±å–®ä¸€ä¸»é¡Œ
topic = "å®¢å»³/message"
qos = 1

result, mid = client.subscribe(topic, qos)
print(f"è¨‚é–±ä¸»é¡Œ: {topic}")
print(f"QoS: {qos}")
print(f"ç­‰å¾…æ¥æ”¶è¨Šæ¯... (å¯ä»¥åŸ·è¡Œ lesson6_1.ipynb çš„ç™¼å¸ƒç¨‹å¼ä¾†æ¸¬è©¦)")
print("æŒ‰ Ctrl+C æˆ–åŸ·è¡Œä¸‹ä¸€å€‹ cell ä¾†åœæ­¢è¨‚é–±")
```

```python
# è¨‚é–±å¤šå€‹ä¸»é¡Œ
topics = [
    ("å®¢å»³/message", 1),
    ("å®¢å»³/sensor", 1),
    ("å®¢å»³/counter", 1)
]

for topic, qos in topics:
    result, mid = client.subscribe(topic, qos)
    print(f"âœ… å·²è¨‚é–±: {topic} (QoS: {qos})")

print("\nç­‰å¾…æ¥æ”¶è¨Šæ¯... (å¯ä»¥åŸ·è¡Œ lesson6_1.ipynb çš„ç™¼å¸ƒç¨‹å¼ä¾†æ¸¬è©¦)")
print("æŒ‰ Ctrl+C æˆ–åŸ·è¡Œä¸‹ä¸€å€‹ cell ä¾†åœæ­¢è¨‚é–±")
```

```python
# è¨‚é–±ä¸»é¡Œä¸¦æŒçºŒç›£è½æŒ‡å®šæ™‚é–“
topic = "å®¢å»³/#"
qos = 1

result, mid = client.subscribe(topic, qos)
print(f"âœ… å·²è¨‚é–±: {topic}")
print("é–‹å§‹ç›£è½è¨Šæ¯...")

# ç›£è½ 30 ç§’ï¼ˆå¯ä»¥ä¿®æ”¹é€™å€‹æ™‚é–“ï¼‰
listen_duration = 30
print(f"å°‡ç›£è½ {listen_duration} ç§’...")

start_time = time.time()
while time.time() - start_time < listen_duration:
    time.sleep(1)

print(f"\nâ° {listen_duration} ç§’ç›£è½æ™‚é–“çµæŸ")
```
