```python
chinese = int (input("國文分數"))
print(type(chinese))
english = int (input ("英文分數"))
print(type(english))
math = int (input("數學分數"))
print(type(math))
total = chinese + english + math
print("總分:",total,"分",sep="-") 
average = round(total/3,ndigits=2)
print("平均:",average,sep="-")
```

```python
#字串差補使用在最前方加上"f"變數用{}隔開，可以讓整體自串與變數結合

chinese = int (input("國文分數"))
print(type(chinese))
english = int (input ("英文分數"))
print(type(english))
math = int (input("數學分數"))
print(type(math))
total = chinese + english + math
print(f"總分:{total}分",sep="-") 
average = round(total/3,ndigits=2)
print(f"平均:{average}分",sep="-")
```

```python
presure =int(input("請輸入血壓"))
if presure >= 90 and presure<=120 :
    print('正常')
else:
    print("不正常")
if presure < 90 or presure>120 :
    print('不正常')
else:
    print("正常")
if not(presure < 90 and presure>120) :
    print('正常')
else:
    print("不正常")
```

```python
import random
five =int(input("請輸入數值"))
NumB = 1 
while random.randint(1,10) != five:
    NumB += 1
    print(f"不等於{five}")
#print(NumB)
print(f"執行第{NumB}次才等於{five}")
```

```python
import random
five =int(input("請輸入數值"))
NumB = 1 
while True:
    if random.randint(1,10) == five:
        break
    NumB += 1
    print(f"不等於{five}")
#print(NumB)
print(f"執行第{NumB}次才等於{five}")
```
